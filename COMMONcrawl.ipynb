1. Ingestion et Prétraitement
Source : fichiers WET de Common Crawl.

Nettoyage HTML : justext ou boilerpy3.

Filtrage de langue : fastText ou langdetect (anglais uniquement).

Filtrage qualité : suppression des documents trop courts ou peu informatifs.


Il existe un dataset “Common Crawl sample” hébergé sur Hugging Face avec un petit sous-ensemble de documents provenant de WET, ce qui évite le téléchargement massif et les erreurs de gzip. Ce dataset est plus fiable pour tester et démarrer.

```python
!pip install datasets -q

from datasets import load_dataset
import json

# On charge un petit échantillon (quelques milliers de docs)
ds = load_dataset("agentlans/common-crawl-sample", split="train")

# Choisir uniquement les documents en anglais
# Le dataset fournit un champ 'language' (géré par gcld3)
english_docs = [ex for ex in ds if ex.get("language") == "en"]
print("Nombre de docs anglais :", len(english_docs))

# Prendre les 20 premiers pour un mini-échantillon
sample = english_docs[:20]

# Sauvegarder en JSONL pour le pipeline
with open("sample_wet_subset.jsonl", "w", encoding="utf-8") as f:
    for ex in sample:
        f.write(json.dumps({"text": ex["text"], "language": ex.get("language")}, ensure_ascii=False) + "\n")

print("Sample JSONL généré avec", len(sample), "documents : sample_wet_subset.jsonl")
```

```python
import json
import re

INPUT_FILE = "sample_wet_subset.jsonl"
OUTPUT_FILE = "cleaned_sample.jsonl"

MIN_WORDS = 100
MIN_UNIQUE_RATIO = 0.3

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # espaces multiples → un espace
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)  # liens, emails
    return text.strip()

def is_quality(text):
    words = re.findall(r'\w+', text.lower())
    if len(words) < MIN_WORDS:
        return False
    unique_ratio = len(set(words)) / len(words)
    return unique_ratio >= MIN_UNIQUE_RATIO

processed_docs = []
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    for line in f:
        doc = json.loads(line)
        if doc.get("language") != "en":
            continue
        cleaned = clean_text(doc["text"])
        if is_quality(cleaned):
            processed_docs.append({"text": cleaned, "language": "en"})

# Sauvegarde
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for doc in processed_docs:
        f.write(json.dumps(doc, ensure_ascii=False) + "\n")

print(f"{len(processed_docs)} documents conservés dans {OUTPUT_FILE}")
```

2. Analyse linguistique
NER avec spaCy ou flair → extraction des entités PERSON, ORG, LOC, etc.

POS tagging pour repérer les verbes et structures relationnelles.

Indicateurs calculés :

Densité d’entités.

Nombre de relations potentielles (entités + verbes).

```python
!pip install -q flair spacy
!python -m spacy download en_core_web_sm

import json
from flair.data import Sentence
from flair.models import SequenceTagger
import spacy

# --- Charger les modèles ---
ner_tagger = SequenceTagger.load("ner")      # Flair NER
nlp = spacy.load("en_core_web_sm")           # spaCy POS

# --- Fichiers ---
INPUT_FILE = "cleaned_sample.jsonl"
OUTPUT_FILE = "analysis_results_with_pos.jsonl"

results = []

# --- Pipeline ---
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    for line in f:
        doc_json = json.loads(line)
        text = doc_json["text"]

        # NER avec Flair
        sentence = Sentence(text)
        ner_tagger.predict(sentence)
        entities = [{"text": ent.text, "label": ent.tag} for ent in sentence.get_spans('ner')]

        # POS tagging avec spaCy
        doc_spacy = nlp(text)
        tokens_with_pos = [{"token": token.text, "pos": token.pos_} for token in doc_spacy]
        verbs = [token["token"] for token in tokens_with_pos if token["pos"] == "VERB"]

        # Indicateurs
        entity_density = len(entities) / len(text.split()) if len(text.split()) > 0 else 0
        potential_relations = len(entities) * len(verbs)

        # Résultat
        results.append({
            "text": text,
            "entities": entities,
            "tokens_pos": tokens_with_pos,   # ajout des tokens avec POS
            "entity_density": entity_density,
            "num_verbs": len(verbs),
            "potential_relations": potential_relations
        })

# --- Sauvegarde ---
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for res in results:
        f.write(json.dumps(res, ensure_ascii=False) + "\n")

print(f"Analyse terminée. {len(results)} documents traités et sauvegardés dans {OUTPUT_FILE}.")
```

3. Labellisation hybride (Weak Supervision)
Heuristiques : densité d’entités + co-occurrence entités/verbes.

Zero-shot classification (facebook/bart-large-mnli) → estimation "relation-rich" / "relation-poor".

Combinaison : vote ou moyenne des scores → pseudo-labels fiables.

```python
import json
import numpy as np
from tqdm import tqdm
from transformers import pipeline

# Paramètres
INPUT_FILE = "analysis_results_with_pos.jsonl"
OUTPUT_FILE = "labeled_results.jsonl"
ENTITY_DENSITY_THRESHOLD = 0.02  # Seuil pour densité d'entités
RELATIONS_THRESHOLD = 30  # Seuil pour co-occurrence entités/verbes
WEIGHT_HEURISTIC = 0.4  # Poids pour score heuristique
WEIGHT_ZERO_SHOT = 0.6  # Poids pour zero-shot

# Initialiser le modèle zero-shot
try:
    classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
except Exception as e:
    print(f"Erreur chargement modèle: {e}. Installez transformers et torch.")
    exit(1)

# Normaliser une valeur entre 0 et 1
def normalize(value, min_val, max_val):
    return (value - min_val) / (max_val - min_val) if max_val > min_val else 0

# Calculer score heuristique
def heuristic_score(entity_density, potential_relations):
    # Normaliser les métriques
    density_norm = normalize(entity_density, 0, 0.1)  # Supposons densité max 0.1
    relations_norm = normalize(potential_relations, 0, 100)  # Supposons max 100 relations
    # Score combiné (moyenne)
    return (density_norm + relations_norm) / 2

# Pipeline principale
def main():
    results = []

    # Lire les données
    with open(INPUT_FILE, "r", encoding="utf-8") as f:
        docs = [json.loads(line) for line in f]

    # Traiter chaque document
    for doc in tqdm(docs, desc="Labellisation"):
        text = doc["text"]
        entity_density = doc["entity_density"]
        potential_relations = doc["potential_relations"]

        # Étape 1 : Score heuristique
        heuristic = heuristic_score(entity_density, potential_relations)
        heuristic_label = "relation-rich" if heuristic > 0.5 else "relation-poor"

        # Étape 2 : Zero-shot classification
        try:
            zero_shot_result = classifier(text[:512], candidate_labels=["relation-rich", "relation-poor"])  # Limiter à 512 tokens
            zero_shot_score = zero_shot_result["scores"][zero_shot_result["labels"].index("relation-rich")]
            zero_shot_label = zero_shot_result["labels"][0]
        except Exception as e:
            print(f"Erreur zero-shot pour texte: {e}")
            zero_shot_score = 0.5  # Score neutre en cas d'erreur
            zero_shot_label = "relation-poor"

        # Étape 3 : Combinaison des scores
        combined_score = WEIGHT_HEURISTIC * heuristic + WEIGHT_ZERO_SHOT * zero_shot_score
        final_label = "relation-rich" if combined_score > 0.5 else "relation-poor"

        # Ajouter le label au document
        doc["heuristic_label"] = heuristic_label
        doc["heuristic_score"] = heuristic
        doc["zero_shot_label"] = zero_shot_label
        doc["zero_shot_score"] = zero_shot_score
        doc["final_label"] = final_label
        doc["combined_score"] = combined_score
        results.append(doc)

    # Sauvegarder
    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        for res in results:
            f.write(json.dumps(res, ensure_ascii=False) + "\n")

    print(f"Labellisation terminée. {len(results)} documents dans {OUTPUT_FILE}")

if __name__ == "__main__":
    main()
```

4. Entraînement auto-supervisé (Contrastive Learning)
Génération automatique de paires :

Positives : documents avec ≥ X entités communes.

Négatives : documents aléatoires sans entités communes.

Modèle : SimCSE ou Sentence-BERT.

Objectif : apprendre des embeddings spécialisés pour capturer la richesse relationnelle.

```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import Counter, defaultdict
import numpy as np
from typing import Dict, List
import warnings
warnings.filterwarnings('ignore')

class ContrastiveLearningAnalyzer:
    """Analyseur pour optimiser le processus de contrastive learning"""

    def __init__(self, data_path: str):
        self.data_path = data_path
        self.documents = []
        self.entity_stats = defaultdict(int)
        self.load_data()

    def load_data(self):
        """Charger et analyser les données"""
        with open(self.data_path, 'r', encoding='utf-8') as f:
            for line_idx, line in enumerate(f):
                doc = json.loads(line.strip())
                entities = [ent['text'] for ent in doc.get('entities', [])]

                self.documents.append({
                    'id': line_idx,
                    'text': doc['text'],
                    'entities': entities,
                    'num_entities': len(entities),
                    'final_label': doc.get('final_label', 'unknown'),
                    'entity_density': doc.get('entity_density', 0),
                    'combined_score': doc.get('combined_score', 0)
                })

                # Compter les entités
                for entity in entities:
                    self.entity_stats[entity] += 1

    def analyze_entity_distribution(self):
        """Analyser la distribution des entités"""
        print("=== ANALYSE DE LA DISTRIBUTION DES ENTITÉS ===")

        # Statistiques générales
        total_docs = len(self.documents)
        total_entities = sum(len(doc['entities']) for doc in self.documents)
        unique_entities = len(self.entity_stats)

        print(f"Nombre total de documents: {total_docs}")
        print(f"Nombre total d'entités: {total_entities}")
        print(f"Nombre d'entités uniques: {unique_entities}")
        print(f"Moyenne d'entités par document: {total_entities/total_docs:.2f}")

        # Distribution du nombre d'entités par document
        entity_counts = [doc['num_entities'] for doc in self.documents]
        print(f"Min entités par doc: {min(entity_counts)}")
        print(f"Max entités par doc: {max(entity_counts)}")
        print(f"Médiane: {np.median(entity_counts):.1f}")

        # Entités les plus fréquentes
        most_common = Counter(self.entity_stats).most_common(10)
        print("\nEntités les plus fréquentes:")
        for entity, count in most_common:
            print(f"  {entity}: {count} occurrences")

        return {
            'total_docs': total_docs,
            'total_entities': total_entities,
            'unique_entities': unique_entities,
            'avg_entities_per_doc': total_entities/total_docs,
            'entity_counts_distribution': entity_counts,
            'most_common_entities': most_common
        }

    def analyze_potential_pairs(self, min_common_entities_range: List[int] = [1, 2, 3, 4, 5]):
        """Analyser le nombre de paires potentielles selon différents seuils"""
        print("\n=== ANALYSE DES PAIRES POTENTIELLES ===")

        results = {}

        for threshold in min_common_entities_range:
            positive_pairs = 0
            entity_to_docs = defaultdict(list)

            # Indexer les documents par entités
            for doc_idx, doc in enumerate(self.documents):
                for entity in doc['entities']:
                    entity_to_docs[entity].append(doc_idx)

            # Compter les paires avec au moins 'threshold' entités communes
            processed_pairs = set()

            for entity, doc_ids in entity_to_docs.items():
                if len(doc_ids) > 1:
                    for i in range(len(doc_ids)):
                        for j in range(i + 1, len(doc_ids)):
                            doc1_id, doc2_id = doc_ids[i], doc_ids[j]

                            pair_key = (min(doc1_id, doc2_id), max(doc1_id, doc2_id))
                            if pair_key in processed_pairs:
                                continue

                            processed_pairs.add(pair_key)

                            # Vérifier le nombre d'entités communes
                            common_entities = set(self.documents[doc1_id]['entities']).intersection(
                                set(self.documents[doc2_id]['entities'])
                            )

                            if len(common_entities) >= threshold:
                                positive_pairs += 1

            results[threshold] = positive_pairs
            print(f"Seuil {threshold} entité(s) commune(s): {positive_pairs} paires positives")

        return results

    def analyze_class_distribution(self):
        """Analyser la distribution des classes de richesse relationnelle"""
        print("\n=== ANALYSE DE LA DISTRIBUTION DES CLASSES ===")

        class_counts = Counter(doc['final_label'] for doc in self.documents)

        for label, count in class_counts.items():
            percentage = (count / len(self.documents)) * 100
            print(f"{label}: {count} documents ({percentage:.1f}%)")

        return class_counts

    def recommend_hyperparameters(self):
        """Recommander des hyperparamètres optimaux"""
        print("\n=== RECOMMANDATIONS D'HYPERPARAMÈTRES ===")

        # Analyser la distribution des entités
        entity_stats = self.analyze_entity_distribution()
        avg_entities = entity_stats['avg_entities_per_doc']

        # Analyser les paires potentielles
        pair_analysis = self.analyze_potential_pairs()

        # Recommandations basées sur l'analyse
        recommendations = {}

        # Seuil d'entités communes
        if avg_entities < 3:
            recommended_threshold = 1
        elif avg_entities < 6:
            recommended_threshold = 2
        else:
            recommended_threshold = 3

        # Vérifier que le seuil génère suffisamment de paires
        pairs_at_threshold = pair_analysis.get(recommended_threshold, 0)
        if pairs_at_threshold < 100:
            recommended_threshold = max(1, recommended_threshold - 1)

        recommendations['min_common_entities'] = recommended_threshold

        # Taille de batch basée sur le nombre de documents
        if len(self.documents) < 1000:
            recommendations['batch_size'] = 8
        elif len(self.documents) < 5000:
            recommendations['batch_size'] = 16
        else:
            recommendations['batch_size'] = 32

        # Nombre d'époques basé sur la taille du dataset
        if len(self.documents) < 1000:
            recommendations['epochs'] = 5
        elif len(self.documents) < 5000:
            recommendations['epochs'] = 3
        else:
            recommendations['epochs'] = 2

        # Ratio de balance positive/négative
        recommendations['balance_ratio'] = 1.0  # 1:1 par défaut

        print(f"Seuil d'entités communes recommandé: {recommendations['min_common_entities']}")
        print(f"Taille de batch recommandée: {recommendations['batch_size']}")
        print(f"Nombre d'époques recommandé: {recommendations['epochs']}")
        print(f"Ratio balance pos/neg recommandé: {recommendations['balance_ratio']}")

        return recommendations

    def visualize_analysis(self, save_plots: bool = True):
        """Créer des visualisations de l'analyse"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))

        # 1. Distribution du nombre d'entités par document
        entity_counts = [doc['num_entities'] for doc in self.documents]
        axes[0, 0].hist(entity_counts, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
        axes[0, 0].set_title('Distribution du nombre d\'entités par document')
        axes[0, 0].set_xlabel('Nombre d\'entités')
        axes[0, 0].set_ylabel('Nombre de documents')

        # 2. Distribution des classes
        class_counts = Counter(doc['final_label'] for doc in self.documents)
        axes[0, 1].bar(class_counts.keys(), class_counts.values(), color=['#FF6B6B', '#4ECDC4'])
        axes[0, 1].set_title('Distribution des classes de richesse relationnelle')
        axes[0, 1].set_xlabel('Classe')
        axes[0, 1].set_ylabel('Nombre de documents')

        # 3. Entités les plus fréquentes (top 15)
        most_common = Counter(self.entity_stats).most_common(15)
        entities, counts = zip(*most_common)
        axes[1, 0].barh(range(len(entities)), counts, color='lightcoral')
        axes[1, 0].set_yticks(range(len(entities)))
        axes[1, 0].set_yticklabels([ent[:20] + '...' if len(ent) > 20 else ent for ent in entities])
        axes[1, 0].set_title('Top 15 des entités les plus fréquentes')
        axes[1, 0].set_xlabel('Fréquence')

        # 4. Relation entre densité d'entités et score combiné
        densities = [doc['entity_density'] for doc in self.documents]
        scores = [doc['combined_score'] for doc in self.documents]
        axes[1, 1].scatter(densities, scores, alpha=0.6, color='purple')
        axes[1, 1].set_title('Densité d\'entités vs Score combiné')
        axes[1, 1].set_xlabel('Densité d\'entités')
        axes[1, 1].set_ylabel('Score combiné')

        plt.tight_layout()

        if save_plots:
            plt.savefig('contrastive_learning_analysis.png', dpi=300, bbox_inches='tight')
            print("\nGraphiques sauvegardés dans 'contrastive_learning_analysis.png'")

        plt.show()

    def export_analysis_report(self, output_file: str = "contrastive_analysis_report.json"):
        """Exporter un rapport complet de l'analyse"""

        entity_stats = self.analyze_entity_distribution()
        pair_analysis = self.analyze_potential_pairs()
        class_distribution = self.analyze_class_distribution()
        recommendations = self.recommend_hyperparameters()

        report = {
            'dataset_summary': {
                'total_documents': len(self.documents),
                'entity_statistics': entity_stats,
                'class_distribution': dict(class_distribution)
            },
            'pair_analysis': pair_analysis,
            'recommendations': recommendations,
            'generated_timestamp': pd.Timestamp.now().isoformat()
        }

        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(report, f, indent=2, ensure_ascii=False)

        print(f"\nRapport d'analyse exporté dans '{output_file}'")
        return report


# Script d'analyse optimisée pour le contrastive learning
def optimize_contrastive_learning(data_path: str):
    """Script principal pour optimiser le contrastive learning"""

    print("🔍 ANALYSE ET OPTIMISATION DU CONTRASTIVE LEARNING")
    print("=" * 60)

    # Initialiser l'analyseur
    analyzer = ContrastiveLearningAnalyzer(data_path)

    # Effectuer l'analyse complète
    analyzer.analyze_entity_distribution()
    analyzer.analyze_potential_pairs()
    analyzer.analyze_class_distribution()
    recommendations = analyzer.recommend_hyperparameters()

    # Créer les visualisations
    analyzer.visualize_analysis()

    # Exporter le rapport
    analyzer.export_analysis_report()

    print("\n" + "=" * 60)
    print("✅ ANALYSE TERMINÉE")
    print("\nProcédure recommandée:")
    print("1. Utilisez les hyperparamètres recommandés ci-dessus")
    print("2. Lancez l'entraînement avec le script principal")
    print("3. Évaluez les résultats et ajustez si nécessaire")

    return recommendations

# Exemple d'utilisation
if __name__ == "__main__":
    # Analyser et optimiser
    recommendations = optimize_contrastive_learning("labeled_results.jsonl")

    # Exemple d'utilisation des recommandations avec le trainer
    print("\n💡 Code d'exemple avec les paramètres optimisés:")
    print(f"""
from contrastive_learning import ContrastiveTrainer

trainer = ContrastiveTrainer("labeled_results.jsonl", min_common_entities={recommendations['min_common_entities']})

model = trainer.train_sentence_bert(
    output_path="./optimized-relation-sbert",
    epochs={recommendations['epochs']},
    batch_size={recommendations['batch_size']}
)
""")
```

```python
import json
import random
import torch
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader

# ======== CONFIG ========
DATA_PATH = "labeled_results.jsonl"
BASE_MODEL = "sentence-transformers/all-MiniLM-L6-v2"
BATCH_SIZE = 16
EPOCHS = 2
ENTITY_THRESHOLD = 1

# ======== DEVICE ========
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Entraînement sur : {device}")

# ======== 1. Charger les données ========
data_list = []
with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        entities = {e["text"] if isinstance(e, dict) and "text" in e else str(e) for e in obj["entities"]}
        data_list.append({"text": obj["text"], "entities": entities})

print(f"{len(data_list)} documents chargés")

# ======== 2. Génération des paires ========
positive_pairs = []
negative_pairs = []

n = len(data_list)
for i in range(n):
    for j in range(i + 1, n):
        common_entities = data_list[i]["entities"].intersection(data_list[j]["entities"])
        if len(common_entities) >= ENTITY_THRESHOLD:
            positive_pairs.append((data_list[i]["text"], data_list[j]["text"]))
        else:
            negative_pairs.append((data_list[i]["text"], data_list[j]["text"]))

print(f"{len(positive_pairs)} paires positives, {len(negative_pairs)} paires négatives")

# ======== 2.1 Rééquilibrage des données ========
if positive_pairs:
    negative_pairs = random.sample(negative_pairs, min(len(negative_pairs), len(positive_pairs)))
else:
    raise ValueError("⚠ Pas de paires positives trouvées. Augmentez ENTITY_THRESHOLD ou vérifiez vos données.")

print(f"Après équilibrage : {len(positive_pairs)} positives, {len(negative_pairs)} négatives")

#gros dataset
##if not positive_pairs:
  #  raise ValueError(" Pas de paires positives trouvées. Augmentez ENTITY_THRESHOLD ou vérifiez vos données.")
# Ici, on ne réduit pas le nombre de négatives##

# ======== 3. Construire le dataset SentenceTransformers ========
train_examples = []
for t1, t2 in positive_pairs:
    train_examples.append(InputExample(texts=[t1, t2], label=1.0))
for t1, t2 in negative_pairs:
    train_examples.append(InputExample(texts=[t1, t2], label=0.0))

random.shuffle(train_examples)

if len(train_examples) == 0:
    raise ValueError("Pas assez de paires pour entraîner le modèle. Vérifie ENTITY_THRESHOLD.")

train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=BATCH_SIZE)

# ======== 4. Charger le modèle sur le bon device ========
model = SentenceTransformer(BASE_MODEL)
model = model.to(device)

# ======== 5. Définir la loss ========
train_loss = losses.CosineSimilarityLoss(model)

# ======== 6. Entraînement ========
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=EPOCHS,
    warmup_steps=100,
    show_progress_bar=True
)

# ======== 7. Sauvegarder le modèle ========
model.save("contrastive_model")
print("Entraînement terminé et modèle sauvegardé dans 'contrastive_model'")
```
